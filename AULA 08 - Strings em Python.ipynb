## **2. STRINGS E LISTAS:**
`Strings` e listas so estruturas compostas por sequncias de dados, cuja ordem  relevante para os processos que as envolvem. A principal distino entre elas  o fato do contedo das strings ser constitudo exclusivamente por texto, enquanto as listas podem ser compostas por qualquer tipo de dado



 |  |STRINGS | LISTAS |
| --------------- |-------------------------------------------------- | --------------------------------------- |
| CONTEDO | Texto | Qualquer tipo de dado |
| MUTABILIDADE | Impossvel de alterar diretamente um caracter | Podemos alterar o valor de um caracter diretamente |
| EXEMPLO | *texto*="baiacu" |*lista*=[10, 'cachorro'] |
| ACESSO POR NDICE| Possvel (EX: texto[0]="b" - acesso por letra) | Possvel (EX: lista[0]=10) |

## **3. ITERAO EM STRINGS:**
**Iterao**  um processo que permite acessar cada caractere de uma string de maneira <b>individual</b>. Geralmente, a iterao utiliza o loop 'for' e, aps cada uma de suas repeties, acessa-se o caracter de ndice posterior ao que foi acessado na repetio atual. A iterao pode ser realizadade duas maneiras diferentes:

<h3><b>EXEMPLO 01:<b></h3>

```python
exemplo01='O Humberto  legal'
for i in range(len(exemplo01)):
  print(exemplo01[i])
```

<br>



<img alt="Strings em Python" src="https://github.com/Ken12345678901/Strings_em_Python/blob/main/img03.jpeg?raw=1" width="520px"></td><td><p style="text-align:justify">

Nesse exemplo, usamos o lao for junto com a funo range(len(exemplo01)). Isso significa que o loop vai percorrer todos os ndices da string exemplo01, um por um.
A cada repetio do for, acessamos um caractere da string usando seu ndice: exemplo01[i]. Como i comea em 0 e vai at o ltimo ndice da string, conseguimos ver ou trabalhar com cada letra individualmente, na ordem em que aparecem.

<h3><b>EXEMPLO 02:<b></h3>

```python
exemplo02='O Humberto  legal'
for elemento in exemplo02:
  print(elemento)
```

## **4. USANDO OPERADORES COM STRINGS:**
Diferentemente de valores numricos (int e float), as strings, mesmo que possuam nmeros em seu contedo, no podem ser utilizadas em clculos. Entretanto, dois smbolos aritmticos: + e * podem manipular as strings, e desempenham, respectivamente, as funes de **concatenao** e **repetio**.

####**CONCATENAO:**

 a unio de duas ou mais strings em uma nica string, cujo contedo  composto pelo texto das strings concatenadas, na ordem em que foram concatenadas. Em Python  efetuada com o smbolo +

<h3><b>EXEMPLO 01:<b></h3>

```python
texto01 = 'Etec Rosa Perrone'
texto02 = ' a melhor escola de Itatiba'
print(texto01+' '+texto02)
```

## **5. EXPLORANDO FUNES COM STRINGS:**


**Funes** so comandos pr-definidos na linguagem python, que permitem manipular estruturas de dados, inclusive as strings



 | NOME DA FUNO | PAPEL |
| ------------ | ---------------------------------------------------- |
| split(x) | Dividir a string em uma lista de palavras ou caracteres, separando-as a cada apario de **x**, o separador |
| x.join() | Junta os elementos de uma lista na formao de uma string nova, em que os elementos esto separados por **x**, chamado de separador|
| lower() | Deixa todas as letras em uma string em minsculo |
| upper() | Deixa todas as letras em uma string em maisculo |
| title() | Deixa todas as letras iniciais de cada palavra da string em maisculo |
| capitalize() | Deixa a letras inicial da primeira palavra da string em maisculo |
| count(x) | Conta quantas vezes o caracter **x** aparece na string |
| startswith(x) | Verifica se uma string comea com uma sequncia de caracteres **x**, retornando True, se comear, e False, se no comear |
| endswith(x) | Verifica se uma string termina com uma sequncia de caracteres **x**, retornando True, se terminar, e False, se no terminar |
| replace(x, y) | Substitu um trecho **x** de uma string por um trecho **y** de caracteres |
| strip() | Remove espaos em branco do incio e do fim da string |

## **6. SLICING EM STRINGS:**


**Slicing**  uma tcnica de manipulao de strings que consiste em selecionar e extrair uma subseo de uma string pr-existente. A estrutura bsica do slicing : `STRING [x:y:z]`, sendo os valores no interior dos colchetes os **parmetros**, ou seja, variveis importantes para a execuo dessa funo.



|PARMETRO| PAPEL | DEFINIO |
| ----------- | --------------------|----------------------------------- |
| x | valor inicial |  o ndice do primeiro elemento da string a ser selecionado (inclusivo) |
| y | valor final |  o ndice do ltimo elemento da string a ser selecionado (exclusivo) |
| z | intervalo |  o intervalo entre os ndices da string. Pode ser um positivo (incremento) ou negativo (decremento)|
